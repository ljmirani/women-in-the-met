#### When were the artworks acquired?

```python
df.groupby('year_acquired')['artistGender'].value_counts()
```

```python
growth = df.groupby(['year_acquired', 'artistGender']).size().unstack(fill_value=0)
```

```python
growth_all_art = growth.cumsum()
```

```python
growth_all_art
```
