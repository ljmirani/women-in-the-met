### 2. From the list of object IDs, make a list of dictionaries with all the Artworks Data and save it 

The MET API will only let you make 80 requests a second, so to make sure I won´t get blocked I put a limit on my requests

```python
import time
min_interval = 1 / 70 
```

```python
all_artists=[]
```

```python
count=0
for ma_id in modern_art_ids:
    start_time = time.time()
    url=f"https://collectionapi.metmuseum.org/public/collection/v1/objects/{ma_id}"
    response = requests.get(url)
    all_artists.append(response.json())
    count=count+1
    print(f"Successfully retrieved data for ID {ma_id}: {count}")
    elapsed_time = time.time() - start_time
    if elapsed_time < min_interval:
        time.sleep(min_interval - elapsed_time)
  
        
        
    
```

### 2. Removing empty/unneeded pieces of data


#### Removing "Unknown Designer" and Empty Strings of Artist Names

```python
df_1 = df.drop(df[(df['artistDisplayName'] == 'Unknown Designer') | (df['artistDisplayName'] == '')].index)
```

```python
df_2=df_1.drop(df_1[
               (df_1['artistRole'] == 'Manufacturer')|
               (df_1['artistRole'] == 'Design House')|
               (df_1['artistRole'] == 'Publisher')
               ].index)
```

### 3. Fixing the accession Year

One problem with my data that I noticed here is that some of the accession year values are missing.  However, while reading the documentation, I noticed that the year the artwork was acquired is also mentioned in the credit line, so what I will try is to extract that year from the column to create a new column with all the acquisition years. If there is still data missing, I will try to merge this column with the AccessionYear column to get as many values as possible.

```python
df_2["accessionYear"].value_counts()
```

```python
df_2.query("accessionYear == ''")
```

#### Trying out the code with a copy of my df, as I do not want to permanently modify my original df if the code does not work

```python
df_2copy = df_2.copy()
```

```python
df_2copy['year_acquired'] = df_2copy['creditLine'].str.extract(r'(\d{4})')
```

```python
df_2copy['year_acquired'].value_counts()
```

```python
df_2copy.info()
```

#### Within this new year_acquired column, 31 values are still missing

```python
#Taking a look at the columns, where the year is missing
df_years_miss=df_2copy[df_2copy['year_acquired'].isnull()]
```

```python
df_years_miss['accessionYear'].value_counts()
```

#### Most columns without a year_acquired value have an accessionYear value. So a (partial) solution to the problem is to merge the accessionYear with the year_acquired column, to reduce the overall number of missing years from 168 (from the original df) to 7.

```python
df_2copy['year_acquired'] = df_2copy['year_acquired'].fillna(df_2copy['accessionYear'])
```
